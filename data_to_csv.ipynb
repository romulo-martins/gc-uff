```python
from github import Github
import pandas as pd
```

```python
# Instancia a blibioteca do Github com a partir de um token de acesso.

g = Github("token")
```

Baixando os reposit√≥rio das linguagens e depois exportando para CSV.

```python
repos = []

lang = "ruby"
result = g.search_repositories(query="language:%s stars:>5000" % lang)
for r in result:
    repo = {
        'id': r.id,
        'name': r.name,
        'full_name': r. full_name,
        'description': r.description,
        'git_url': r.git_url,
        'commits': r.get_commits().totalCount,
        'subscribers_count': r.subscribers_count,
        'stargazers_count': r.stargazers_count,
        'watchers_count': r.watchers_count,
        'open_issues_count': r.open_issues_count,
        'forks_count': r.forks_count,
        'master_branch': r.master_branch,
        'language': r.language,
        'size': r.size
    }
    repos.append(repo)
```

```python
df = pd.DataFrame(repos)
df.to_csv('repos-%s.csv' % lang)
```
