Setting PyGithub

```python
from github import Github

g = Github("your token")
```

```python
lang = 'JavaScript'
```

Search for repositories that use a specific programming language and has more than 5 thousand stars

```python
result = g.search_repositories(query="language:{0} stars:>5000 is:public".format(lang))
```

Function to access coveralls API

```python
import requests

def get_test_coverage(full_name):
    try:
        resp = requests.get('https://coveralls.io/github/{0}.json'.format(full_name))
        return resp.json()['covered_percent']
    except:
        return None

# get_test_coverage('cucumber/cucumber-ruby')
```

Iterates 
Itera sobre os reposit√≥rios e busca a cobertura de c√≥digo

```python
repos = []
for r in result:
    repo = {
        'id': r.id,
        'name': r.name,
        'full_name': r.full_name,
        'description': r.description,
        'stargazers_count': r.stargazers_count,
        'open_issues_count': r.open_issues_count,
        'language': r.language,
        'test_coverage': get_test_coverage(r.full_name)
    }
    repos.append(repo)
```

```python
import pandas as pd

df = pd.DataFrame(repos)
df
```
