Configurando o PyGithub

```python
from github import Github

g = Github("token")
```

```python
lang = 'javascript'
```

Filtrando os repositÃ³rios com mais de 5 mil estrelas

```python
result = g.search_repositories(query="language:{0} stars:>5000 is:public".format(lang))
```

```python
import requests
import pandas as pd
from pandas.io.json import json_normalize

def get_test_coverage(full_name):
    try:
        resp = requests.get('https://codecov.io/api/gh/{0}'.format(full_name))
        df = json_normalize(resp.json()['commits'])
        df = df[df['timestamp'] == df['timestamp'].max()]
        return df['totals.c'][0]
    except:
        return None
    
# get_test_coverage('scikit-learn/scikit-learn')
```

Itera sobre os repositÃ³rios e busca a cobertura de testes

```python
repos = []
for r in result:
    repo = {
        'id': r.id,
        'name': r.name,
        'full_name': r.full_name,
        'description': r.description,
        'stargazers_count': r.stargazers_count,
        'open_issues_count': r.open_issues_count,
        'language': r.language,
        'size': r.size,
        'test_coverage': get_test_coverage(r.full_name)
    }
    repos.append(repo)
```

```python
import pandas as pd

df = pd.DataFrame(repos)
df
```

Total de repositorios com mais de 5 mil estrelas

```python
len(repos)
```

Total de repositÃ³rios com coberturas feitas pelo Codecov

```python
len(df[df['test_coverage'].notnull()])
```

Salvando em CSV
